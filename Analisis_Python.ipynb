# **Ejercicio 6**

Use t茅cnicas de normalizaci贸n de datos y grafique para obtener un resultado similar al gr谩fico
siguiente:

```python
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns

df = pd.read_csv('TUNJA_I_23.csv', encoding='ISO-8859-1', delimiter=';', decimal=',')

# Definir las columnas a normalizar
columnas = ['Temperatura', 'Velocidad del Viento', 'Direcci贸n del Viento', 'Presi贸n', 'Punto de Roc铆o', 'Cobertura total nubosa', 'Humedad', 'Pron贸stico ']

# Codificar la columna "Pron贸stico" con valores num茅ricos
codificacion = {'Nublado': 0, 'Nublado - Llovizna': 1, 'Nublado - Lluvia': 2}
df['Pron贸stico '] = df['Pron贸stico '].map(codificacion)

# Reemplazar comas por puntos en las columnas num茅ricas
df[columnas] = df[columnas].replace(',', '.', regex=True)

# Convertir las columnas a tipo float
df[columnas] = df[columnas].astype(float)

# Normalizar los datos de las columnas seleccionadas
scaler = MinMaxScaler()
df_normalizado = pd.DataFrame(scaler.fit_transform(df[columnas]), columns=columnas)

# Generar gr谩ficas de datos normalizados para cada columna
for columna in columnas:
    plt.figure()
    sns.lineplot(data=df_normalizado[columna])
    plt.ylabel(columna)
    plt.xlabel('ndice')
    plt.show()
```
